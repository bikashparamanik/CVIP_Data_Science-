```python
print('Country with most attacks: ',df['Country'].value_counts().idxmax())
print('City with most attacks: ',df['city'].value_counts().index[1])
print("Region with the most attacks:",df['Region'].value_counts().idxmax())
print("Year with the most attacks:",df['Year'].value_counts().idxmax())
print("Month with the most attacks:",df['Month'].value_counts().idxmax())
print("Group with the most attacks:",df['Group'].value_counts().index[1])
print("Most Attack Types:",df['AttackType'].value_counts().idxmax())
```

```python
#Barplot
import seaborn as sns

x_year = df['Year'].unique()
y_year = df['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(15,10))
plt.title("Attack in Years")
plt.xlabel("Attack Years")
plt.ylabel("Number of attacks each year")
plt.xticks(rotation=45)
sns.barplot(x=x_year, y=y_year, palette= 'rocket')
plt.show()
```

```python

#Count plot
plt.figure(figsize=(15, 6))
sns.countplot(x='Year', data=df, palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 5))
plt.xticks(rotation=45)
plt.title('Count Plot of Number Of Terrorist Activities Each Year')
plt.show()
```

```python
#Area plot
pd.crosstab(df.Year,df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist activites by Region in each year')
plt.ylabel("Number of attacks")
plt.show()
```

```python
from wordcloud import WordCloud
from scipy import signal
cities=df.state.dropna(False)
plt.subplots(figsize=(10,10))
wordcloud= WordCloud(background_color='white', width=500, height = 250).generate(' '.join(cities))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['casualities'] = df['Killed'] + df['Wounded']
```

```python
#Taking only 30 countries for visualization of heatmap properly
terror_copy = df.sort_values(by='casualities',ascending=False)[:30]
```

```python
terror_copy.corr()
```

```python
#Generating heatmap for correlation ploy
plt.figure(figsize=[15,8])
sns.heatmap(terror_copy.corr(),cmap='RdYlGn', linewidths=0.4)
plt.show()
```

```python
attack_country = df.Country.value_counts()[:15]
attack_country
```

```python
## Countries affected by terrorism
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_country.index, y=attack_country.values, palette='Blues_d')
plt.title('Top Countries Affected by Terrorism')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_groups = df.Group.value_counts()[1:15]
attack_groups
```

```python
## Groups involving most in terrorism
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_groups.index, y=attack_groups.values, palette="hls")
plt.title('Groups Involved Most in Terrorism')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.xticks(rotation=50)
plt.show()
```

```python
#Terrorist group names and killed each year
group_killed= df[['Group','Killed']].groupby(['Group'],axis=0).sum().sort_values('Killed', ascending=False).head(15)
group_killed
```

```python
## People Killed by each group in terrorist activity
plt.figure(figsize=(12, 6))
sns.barplot(x=group_killed.index, y=group_killed['Killed'], palette="crest")
plt.title('Number of People Killed by Each Terrorist Group')
plt.xlabel('Terrorist Groups')
plt.ylabel('Number of People Killed')
plt.xticks(rotation=45)
plt.show()
```

```python
died_people = df[['Year', 'Killed']].groupby(['Year']).sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=died_people.index, y=died_people['Killed'], palette="viridis")
plt.title("Number of People Died in Each Year", fontsize=15)
plt.ylabel("Number of People Died")
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_cities = df.city.value_counts()[:15]
attack_cities
```

```python
## Top Countries affected
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_cities.index, y=attack_cities.values, palette="viridis")
plt.title('Top Countries Affected')
plt.xlabel('Cities')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_type = df.AttackType.value_counts()[:15]
attack_type
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_type.index, y=attack_type.values, palette="Set3")
plt.title('Attack Types')
plt.xlabel('Types of Attacks')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_killed = df[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum()
attack_killed
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_killed.index, y=attack_killed['Killed'], palette="husl")
plt.title('Number of People Killed in Each Attack Type')
plt.xlabel('Types of Attacks')
plt.ylabel('Number of People Killed')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_wounded = df[['AttackType','Wounded']].groupby(["AttackType"],axis=0).sum()
attack_wounded
```

```python
## People Wounded in each attack type
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_wounded.index, y=attack_wounded['Wounded'], palette="Paired")
plt.title('Number of People Wounded in Each Attack Type')
plt.xlabel('Types of Attacks')
plt.ylabel('Number of People Wounded')
plt.xticks(rotation=45)
plt.show()
```

```python
#Time Series Plot of Terrorist Incidents Over Years:
incident_counts = df['Year'].value_counts().reset_index().rename(columns={'index': 'Year', 'Year': 'Incidents'})

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Incidents', data=incident_counts, marker='o', color=sns.color_palette("Set2")[0])
plt.title('Terrorist Incidents Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()
```
